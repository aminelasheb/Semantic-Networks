# Réseaux sémantiques, Semantic networks

importing the NetworkX library and creating an empty graph using the nx.Graph() function

NetworkX is a powerful library for creating, manipulating, and studying the structure, dynamics, and functions of complex networks. It provides tools for working with graphs, including functions for adding and removing nodes and edges, querying the properties of nodes and edges, and calculating graph-theoretic properties such as centrality and connectivity. It is open-source and actively maintained, making it a popular choice among data scientists and researchers.

```python
import networkx as nx

G = nx.Graph()
```

Then add nodes to the graph using the add_nodes_from() function, representing different types of birds

```python
G.add_nodes_from(["Titi", "Moineau", "Oiseau", "Ailles", "Espèce en danger", "Scientifiques"])
```

After that, use the add_edge() function to add edges to the graph, representing relationships between the birds.

```python
G.add_edge("Titi", "Moineau", relationship = "Est un")
G.add_edge("Moineau", "Oiseau", relationship = "Sorte de")
G.add_edge("Oiseau", "Ailles", relationship = "a")
G.add_edge("Moineau", "Espèce en danger", relationship = "Est un")
G.add_edge("Espèce en danger", "Scientifiques", relationship = "Etudiée par")
```

The edges have an attribute "relationship" that describes the type of relationship between the birds

Prints the edges of the graph to view the relationships that are represented in the network : 

```python
print(G.edges(data=True))
```

using matplotlib and NetworkX to draw and visualize the semantic network that I created earlier.



```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G,'relationship')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()
```
